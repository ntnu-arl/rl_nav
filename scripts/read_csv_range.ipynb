```python
import numpy as np
import matplotlib.pyplot as plt
filename = "./bins.csv"
data = np.loadtxt(filename, delimiter=",")
data[data > 10] = 10
# make the y axis appear longer: taller figure and stretch y via aspect
fig, ax = plt.subplots()  # (width, height)
im = ax.imshow(data, cmap="plasma", aspect=5)  # aspect>1 stretches y
ax.set_title("After Downsampling")
plt.colorbar(im, ax=ax)
plt.show()

filename = "./bins_downsampled.csv"
data = np.loadtxt(filename, delimiter=",")
plt.imshow(data, cmap="plasma")
plt.title("After Downsampling")
plt.colorbar()
plt.show()

# show minimum and maximum values of bins_downsampled.csv
print("Min value:", np.min(data))
print("Max value:", np.max(data))


 
# filename = "./points3d_xyz.csv"
# data = np.loadtxt(filename, delimiter=",")
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.scatter(data[:,0], data[:,1], data[:,2], s=1)
# ax.set_title("3D Points")
# plt.show()

```
